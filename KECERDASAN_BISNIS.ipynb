```python
df.drop_duplicates(inplace=True)
```

```python
# Standarisasi teks
df['Region'] = df ['Region'].str.strip().str.title()
df['Product_Category'] = df['Product_Category'].str.strip().str.title()
df['Product_Name'] = df['Product_Name'].str.strip().str.title()
```

```python
# Mengisi Nilai  Quantity jika memungkinkan (Total Sales dan Unit Price Ada)
df.loc[
    (df['Quantity'].isnull()) &
    (df['Unit_Price'].notnull()) &
    (df['Total_Sales'].notnull()),
    'Quantity'
] = df['Total_Sales'] / df['Unit_Price']


# Mengisi Nilai  Unit_Price jika memungkinkan (Total Sales dan Quantity ada)
df.loc[
    (df['Unit_Price'].isnull()) &
    (df['Quantity'].notnull()) &
    (df['Total_Sales'].notnull()),
    'Unit_Price'
] = df['Total_Sales'] / df['Quantity']


# Jika masih kosong → isi median
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].median())



# Numerik diisi median (nilai tengah)
#df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
#df['Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].median())


# Kategori diisi modus (nilai sering muncul)
df['Product_Name'] = df.groupby('Product_Category')['Product_Name']\
  .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))
```

```python
# Untuk menghitung total sales
df['Total_Sales'] = df['Quantity'] * df['Unit_Price']
```

```python
# Mengubah tipe data
df['Quantity'] = df['Quantity'].astype(int)
df['Unit_Price'] = df['Unit_Price'].astype(float)
df['Total_Sales'] = df['Total_Sales'].astype(float)
```

```python
# KONVERSI ORDER DATE
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', errors='coerce')
```
